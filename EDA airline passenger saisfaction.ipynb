Analyze 'Gender' column

```python
df['Gender'].describe()
```

```python
df['Gender'].value_counts()
```

```python
gender_counts = df['Gender'].value_counts() 
colors = ('pink', 'lightblue')
ax = gender_counts.plot(kind = 'bar', title = 'Gender Categories', color = colors)

for i, count in enumerate(gender_counts):
    ax.text(i, count, str(count), ha='center', va= 'bottom')

ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha= 'center')
ax.set_xlabel('')

plt.show() 
```

```python
df['Age'].describe()
```

```python
age_data = df['Age']

plt.figure(figsize=(14, 6))
plt.hist(age_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
male_counts=df[df['Gender']=='Male']['Age']
plt.figure(figsize=(14,6))
plt.hist(male_counts, bins=30, color='skyblue', edgecolor='black' )
plt.xlabel('Age')
plt.title('Male Counts by Age')
plt.show()
```

```python
female_counts=df[df['Gender']=='Female']['Age']

plt.figure(figsize=(14,6))
plt.hist(female_counts, bins=30, color='lightpink', edgecolor='black')
plt.xlabel('Age')
plt.title('Female Counts by Age')
plt.show()
```

```python
x = df.groupby('Gender')['Age'].value_counts().sort_index()
female_counts = x['Female']
male_counts = x['Male']
```

```python
plt.figure(figsize=(20,20))
male_counts.plot(kind='bar', color='lightblue')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Counts of Ages for Males')
plt.show()

plt.figure(figsize=(20,20))
female_counts.plot(kind='bar', color='lightpink')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Counts of Age for Females')
plt.show()
```

```python
class_count = df['Class'].value_counts()
plt.figure(figsize=(10,6))
colors=['lightpink', 'lightblue','silver' ]
class_count.plot(kind='bar', color=colors)
plt.xticks(rotation=0)
plt.title('Classes')
plt.xlabel('')
for i, count in enumerate(class_count):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.show()
```

```python
travel_types= df['Type_of_Travel'].value_counts()

plt.figure(figsize=(10,6))
colors = ['lightpink', 'lightblue']
travel_types.plot(kind='bar', color=colors)
plt.xticks(rotation=0)
plt.xlabel('')
plt.title('Types of Travel')
for i, count in enumerate(travel_types):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.show()
```
